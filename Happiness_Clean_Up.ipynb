```python
import pandas as pd
from sqlalchemy import create_engine
from config import password
```

```python
countries_df = pd.read_csv("country_abbreviations2.csv")

countries = countries_df['Name'].to_list()
```

```python
new_names = [country.split(',')[0] for country in countries]
new_names
```

```python
countries_df["new_name"] = new_names
countries_df = countries_df.drop(columns = ["Name"])
countries_df = countries_df.rename(columns = {"Code": "code"})
countries_df
```

```python
# Read in file
happy15_df = pd.read_csv("2015_worldhappiness.csv")

# Drop unnecessary columns
happy15_df = happy15_df.drop(columns = ["Region", "Standard Error", "Dystopia Residual"])

# Rename columns
happy15_df = happy15_df.rename(columns = {"Country": "country",
                            "Happiness Rank": "happiness_rank",
                            "Happiness Score": "happiness_score",
                            "Economy (GDP per Capita)": "economy_gdp",
                            "Family": "social_support",
                            "Health (Life Expectancy)": "life_expectancy",
                            "Freedom": "freedom",
                            "Trust (Government Corruption)": "govt_corruption",
                            "Generosity": "generosity"})

# Create year column for 2015
happy15_df["year_recorded"] = "2015"

happy15_df
```

```python
# Merge with country name & abbreviation dataset
happy15_country = happy15_df.merge(countries_df, how='left', left_on='country', right_on='new_name')
happy15_country
```

```python
# Find where Name is NaN and clean up country abbreviation file as needed
is_NaN = happy15_country.isnull()
row_has_NaN = is_NaN.any(axis=1)
rows_with_NaN = happy15_country[row_has_NaN]
rows_with_NaN
```

```python
happy15_country = happy15_country[["year_recorded", "code", "country", "happiness_rank", 
                                   "happiness_score", "economy_gdp", "social_support", 
                                   "life_expectancy", "freedom", "govt_corruption", "generosity"]]
happy15_country
```
