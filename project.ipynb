```python
import pandas as pd
```

```python
# Study data files
femalePop_metadata = "data/Female_population.csv"
femaleLabor_metadata = "data/Female_labor.csv"
femaleEd_metadata = "data/Female_education.csv"
country_metadata = "data/country_abbreviations.csv"

# Read the data
femalePop_metadata = pd.read_csv(femalePop_metadata)
femaleLabor_metadata = pd.read_csv(femaleLabor_metadata)
femaleEd_metadata = pd.read_csv(femaleEd_metadata)
country_metadata = pd.read_csv(country_metadata)
```

```python
country_metadata
```

```python
femaleLabor_metadata
```

```python
femaleLabor_dropped = femaleLabor_metadata.drop(columns=["Series Code", "1990 [YR1990]","2000 [YR2000]", "2011 [YR2011]","2012 [YR2012]", "2013 [YR2013]","2014 [YR2014]","2020 [YR2020]"])
femaleLabor_dropped
```

```python
femaleLabor_merged = femaleLabor_dropped.merge(country_metadata,left_on="Country Name", right_on="Name", how="left")
femaleLabor_merged
```

```python
femaleLabor_na = femaleLabor_merged.dropna(subset=["Name"])
```

```python
femaleLabor_dup = femaleLabor_na.drop_duplicates(subset=["Country Name"])
femaleLabor_dup
```

```python
new_order = ["Country Name","Code","Series Name","2015 [YR2015]","2016 [YR2016]","2017 [YR2017]", "2018 [YR2018]","2019 [YR2019]"]
femaleLabor_df = femaleLabor_dup.reindex(columns=new_order)
femaleLabor_df_2015 = femaleLabor_df[["Country Name","Code","Series Name","2015 [YR2015]"]]
femaleLabor_df_2015 = femaleLabor_df_2015.rename(columns={"2015 [YR2015]":"2015", "Country Name": "Country"})
femaleLabor_df_2015

femaleLabor_df_2016 = femaleLabor_df[["Country Name","Code","Series Name","2016 [YR2016]"]]
femaleLabor_df_2016 = femaleLabor_df_2016.rename(columns={"2016 [YR2016]":"2016","Country Name": "Country"})

femaleLabor_df_2017 = femaleLabor_df[["Country Name","Code","Series Name","2017 [YR2017]"]]
femaleLabor_df_2017 = femaleLabor_df_2017.rename(columns={"2017 [YR2017]":"2017","Country Name": "Country"})

femaleLabor_df_2018 = femaleLabor_df[["Country Name","Code","Series Name","2018 [YR2018]"]]
femaleLabor_df_2018 = femaleLabor_df_2018.rename(columns={"2018 [YR2018]":"2018","Country Name": "Country"})

femaleLabor_df_2019 = femaleLabor_df[["Country Name","Code","Series Name","2019 [YR2019]"]]
femaleLabor_df_2019 = femaleLabor_df_2019.rename(columns={"2019 [YR2019]":"2019","Country Name": "Country"})
```

```python
# Female education 
femaleEd_metadata
```

```python
femaleEd_dropped = femaleEd_metadata.drop(columns=["1990","2000", "2011","2012", "2013","2014","Unnamed: 11","2020"])
femaleEd_dropped
```

```python
femaleEd_merged = femaleEd_dropped.merge(country_metadata,left_on=" Country", right_on="Name", how="left")
femaleEd_merged
```

```python
femaleEd_na = femaleEd_merged.dropna(subset=["Name"])
```

```python
femaleEd_dup = femaleEd_na.drop_duplicates(subset=[" Country"])
femaleEd_dup = femaleEd_dup.drop(columns=["Code"])
femaleEd_dup
```

```python
new_order = [" Country","Code","2015","2016","2017", "2018","2019"]
femaleEd_df = femaleEd_dup.reindex(columns=new_order)


femaleEd_df_2015 = femaleEd_df[[" Country","Code","2015"]]
femaleEd_df_2015 = femaleEd_df_2015.rename(columns={" Country":"Country"})
femaleEd_df_2015

femaleEd_df_2016 = femaleEd_df[[" Country","Code","2016"]]
femaleEd_df_2016 = femaleEd_df_2016.rename(columns={" Country":"Country"})

femaleEd_df_2017 = femaleEd_df[[" Country","Code","2017"]]
femaleEd_df_2017 = femaleEd_df_2017.rename(columns={" Country":"Country"})

femaleEd_df_2018 = femaleEd_df[[" Country","Code","2018"]]
femaleEd_df_2018 = femaleEd_df_2018.rename(columns={" Country":"Country"})

femaleEd_df_2019 = femaleEd_df[[" Country","Code","2019"]]
femaleEd_df_2019 = femaleEd_df_2019.rename(columns={" Country":"Country"})
```

```python
#female population data 
femalePop_metadata 
```

```python
femalePop_dropped = femalePop_metadata.drop(columns=["1990","2000", "2011","2012", "2013","2014","Unnamed: 12","2020"])
femalePop_dropped
```

```python
femalePop_merged = femalePop_dropped.merge(country_metadata,left_on=" ", right_on="Name", how="left")
femalePop_merged
```

```python
femalePop_na = femalePop_merged.dropna(subset=["Name"])
```

```python
femalePop_dup = femalePop_na.drop_duplicates(subset=[" "])
femalePop_dup
```

```python
new_order = [" ","Code","2015","2016","2017", "2018","2019"]
femalePop_df = femalePop_dup.reindex(columns=new_order)


femalePop_df_2015 = femalePop_df[[" ","Code","2015"]]
femalePop_df_2015 = femalePop_df_2015.rename(columns={" ":"Country"})
femalePop_df_2015

femalePop_df_2016 = femalePop_df[[" ","Code","2016"]]
femalePop_df_2016 = femalePop_df_2016.rename(columns={" ":"Country"})

femalePop_df_2017 = femalePop_df[[" ","Code","2017"]]
femalePop_df_2017 = femalePop_df_2017.rename(columns={" ":"Country"})

femalePop_df_2018 = femalePop_df[[" ","Code","2018"]]
femalePop_df_2018 = femalePop_df_2018.rename(columns={" ":"Country"})

femalePop_df_2019 = femalePop_df[[" ","Code","2019"]]
femalePop_df_2019 = femalePop_df_2019.rename(columns={" ":"Country"})
```

```python
df2015_outer = pd.merge(femaleLabor_df_2015, femaleEd_df_2015, on="Country", how='outer')
df2015_outer = pd.merge(df2015_outer, femalePop_df_2015, on="Country", how='outer')
df2015_outer["Year"]="2015"

df2015_outer = df2015_outer.rename(columns={"Code_x":"Code", "2015_x": "female_labor", "2015_y": "female_Education","2015": "female_population"})
df2015_outer= df2015_outer.drop(columns=["Code_y"])
df2015_outer
```

```python
df2016_outer = pd.merge(femaleLabor_df_2016, femaleEd_df_2016, on="Country", how='outer')
df2016_outer = pd.merge(df2016_outer, femalePop_df_2016, on="Country", how='outer')
df2016_outer["Year"]="2016"

df2016_outer = df2016_outer.rename(columns={"Code_x":"Code", "2016_x": "female_labor", "2016_y": "female_Education","2016": "female_population"})
df2016_outer = df2016_outer.drop(columns=["Code_y"])
df2016_outer
```

```python
df2017_outer = pd.merge(femaleLabor_df_2017, femaleEd_df_2017, on="Country", how='outer')
df2017_outer = pd.merge(df2017_outer, femalePop_df_2017, on="Country", how='outer')
df2017_outer["Year"]="2017"

df2017_outer = df2017_outer.rename(columns={"Code_x":"Code", "2017_x": "female_labor", "2017_y": "female_Education","2017": "female_population"})
df2017_outer= df2017_outer .drop(columns=["Code_y"])
df2017_outer
```

```python
df2018_outer = pd.merge(femaleLabor_df_2018, femaleEd_df_2018, on="Country", how='outer')
df2018_outer = pd.merge(df2018_outer, femalePop_df_2018, on="Country", how='outer')
df2018_outer["Year"]="2018"

df2018_outer = df2018_outer.rename(columns={"Code_x":"Code", "2018_x": "female_labor", "2018_y": "female_Education","2018": "female_population"})
df2018_outer= df2018_outer .drop(columns=["Code_y"])
df2018_outer
```

```python
df2019_outer = pd.merge(femaleLabor_df_2019, femaleEd_df_2019, on="Country", how='outer')
df2019_outer = pd.merge(df2019_outer, femalePop_df_2019, on="Country", how='outer')
df2019_outer["Year"]="2019"

df2019_outer = df2019_outer.rename(columns={"Code_x":"Code", "2019_x": "female_labor", "2019_y": "female_Education","2019": "female_population"})
df2019_outer= df2019_outer .drop(columns=["Code_y"])
df2019_outer
```

```python
df_concat = pd.concat([df2015_outer, df2016_outer,df2017_outer,df2018_outer,df2019_outer])
df_concat
```

```python
df_concat.to_csv("female_world.csv")
```
